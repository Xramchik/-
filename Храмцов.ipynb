```python
#Сгенерировать случайным образом неориентированный граф,
#число вершин равно 100+ Свой номер в списке для Контрольной работы.
from collections import deque
import itertools
import networkx as nx
import numpy.random as rnd
import matplotlib.pyplot as plt


graph = nx.Graph()

i = 120
while int(i) > -1:
    graph.add_node(i)
    i = int(i) - 1
graph.nodes()

def add_edge(f_item, s_item, graph=None):
    graph.add_edge(f_item, s_item)
    graph.add_edge(s_item, f_item) 

y = 120
while y > 0:
    add_edge(y, y-1, graph=graph)
    y = y - 1
add_edge(0, 120, graph=graph)
nx.draw_circular(graph,
         node_color='w',
         node_size=1,
         with_labels=True)
```

```python
  #Перенумеровать вершины графа в порядке обхода в ширину
prepared1 = []
pile = deque()

def bfs(graph, begin=0):
    global pile, prepared1
    pile.append(begin)
    while pile:
        c = pile.popleft()
        if not c in prepared1:
            prepared1.append(c)
            print(c)
            for i in graph[c]:
                if not i in prepared1:
                    pile.append(i)
            bfs(graph, pile.pop())


bfs(graph)                                        #Перенумеровать вершины графа в порядке обхода в ширину
```

```python
#Перенумеровать вершины графа в порядке обхода в глубину
prepared = []
pile = deque()


def dfs(graph, begin=0):
    global pile, prepared
    pile.append(begin)
    while pile:
        x = pile.pop()
        if not x in prepared:
            prepared.append(x)
            print(x)
            for i in graph[x]:
                if not i in prepared:
                    pile.append(i)
            dfs(graph, pile.pop())


dfs(graph)                                
```
